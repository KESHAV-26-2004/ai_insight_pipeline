```python
import pandas as pd
import os
import sys

# Force UTF-8 to avoid emoji issues
try:
    sys.stdout.reconfigure(encoding='utf-8')
except:
    pass

# Import pipeline
sys.path.append(os.path.abspath("./"))
from pipeline.runner import run_pipeline

# -----------------------------
# STEP 1 — Ask user for file path
# -----------------------------
print("Enter the FULL PATH to your CSV file:")
print("Example: C:/Users/keshu/Desktop/mydata.csv\n")

csv_path = input("CSV Path: ").strip()

while not os.path.exists(csv_path):
    print("❌ File not found. Try again.")
    csv_path = input("CSV Path: ").strip()

print(f"\n📁 File selected: {csv_path}")

# Load preview
df = pd.read_csv(csv_path, on_bad_lines='skip', low_memory=False)
print("\n===== PREVIEW =====")
display(df.head())

columns = list(df.columns)
print("\nDetected columns:", columns)

# -----------------------------
# STEP 2 — Ask targets
# -----------------------------
primary_target = input("\nEnter PRIMARY target column: ").strip()
while primary_target not in columns:
    primary_target = input("Invalid. Enter PRIMARY column again: ").strip()

secondary_raw = input("\nEnter up to 2 SECONDARY columns (comma separated): ").strip()
secondary_targets = [s.strip() for s in secondary_raw.split(",") if s.strip()][:2]

for s in secondary_targets:
    if s not in columns:
        raise ValueError(f"Invalid secondary target: {s}")

if primary_target in secondary_targets:
    raise ValueError("Primary cannot also be secondary.")

task_type = input("\nEnter task type (classification / regression): ").strip().lower()
while task_type not in ["classification", "regression"]:
    task_type = input("Invalid. Enter classification or regression: ").strip().lower()

# -----------------------------
# STEP 3 — Run pipeline
# -----------------------------
print("\n\n==============================")
print("🚀 Running Full Pipeline")
print("==============================\n")

run_pipeline(
    csv_path,
    primary_target=primary_target,
    secondary_targets=secondary_targets,
    task_type=task_type
)

print("\n🎉 DONE — Check your results/ folder!")
```
